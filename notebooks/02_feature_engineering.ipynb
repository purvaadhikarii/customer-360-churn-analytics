```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/preeti adhikari/OneDrive/Purva Adhikari/OneDrive/Desktop/customer-360-churn-analytics/data/raw/telco_customer_churn_synthetic.csv")

# Handle missing values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

# Encode categorical columns
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in df.select_dtypes(include='object'):
    if col != 'customerID':
        df[col] = encoder.fit_transform(df[col])

# Feature engineering
df['tenure_group'] = pd.cut(df['tenure'],
                            bins=[0,12,24,48,72],
                            labels=['0-1yr','1-2yr','2-4yr','4-6yr'])

df['charges_ratio'] = (df['MonthlyCharges'] / (df['TotalCharges'] + 1)).round(3)

df.to_csv("C:/Users/preeti adhikari/OneDrive/Purva Adhikari/OneDrive/Desktop/customer-360-churn-analytics/data/processed/processed_telco.csv", index=False)

df.head()
```
